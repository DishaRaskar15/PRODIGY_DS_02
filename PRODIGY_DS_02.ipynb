```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df_test = pd.read_csv("C:/Users/Admin/Downloads/test.csv")
```

```python
df_test.columns = df_test.columns.str.strip().str.lower()
```

```python
num_cols = df_test.select_dtypes(include='number').columns
df_test[num_cols] = df_test[num_cols].fillna(df_test[num_cols].median())
```

```python
cat_cols = df_test.select_dtypes(include='object').columns
df_test[cat_cols] = df_test[cat_cols].fillna(df_test[cat_cols].mode().iloc[0])
```

```python
sns.countplot(x='overallqual', data=df_test)
plt.title("Overall Quality Distribution")
plt.xlabel("Overall Quality")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='garagecars', data=df_test)
plt.title("Garage Cars Distribution")
plt.xlabel("Garage Cars")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df_test['lotarea'], bins=30, kde=True, color='purple')
plt.title("Lot Area Distribution")
plt.xlabel("Lot Area")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(df_test['yearbuilt'], bins=30, kde=False, color='orange')
plt.title("Year Built Distribution")
plt.xlabel("Year Built")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df_test['grlivarea'], bins=30, kde=True, color='green')
plt.title("Above Ground Living Area Distribution")
plt.xlabel("GrLivArea")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='garagecars', y='lotarea', data=df_test)
plt.title("Lot Area by Number of Garage Cars")
plt.xlabel("Garage Cars")
plt.ylabel("Lot Area")
plt.show()
```

```python
sns.boxplot(x='overallqual', y='grlivarea', data=df_test)  # simplified, warning-free
plt.title("GrLivArea by Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("GrLivArea")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = df_test.corr(numeric_only=True)
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False, linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```
